```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt


d       = scipy.io.loadmat('spike_trains.mat',squeeze_me=True)
Xa      = d['Xa']
Xb      = d['Xb']
t_data  = np.arange(Xa.size) # time points of the spikes time series in arbitrary units

def kernel_Gaussian(t, W):
    return np.exp(-0.5 * (t / W)**2) / (np.sqrt(2*np.pi) * W)

def kernel_Exp(t, W):
    return np.heaviside(t,1) * np.exp(-t/W) / W

def kernel_Rect(t, W):
    return (np.heaviside(t+W/2,1)-np.heaviside(t-W/2,1)) / W
```

```python
"""

Change these parameters

Test:
* different Kernels
* different W

a) Can you identify the stimulus that generated the spike train?
b) Can you tell whether any of the two neurons was 'noisier' than the other?

"""

W        = 1500                         # width of time bin
t_kernel = np.arange(-4*W,4*W)          # time domain where the kernel is defined
K        = kernel_Gaussian(t_kernel,W)  # kernel curve

"""
don't need to modify below
"""

rho1     = np.convolve(Xa,K,'same')
rho2     = np.convolve(Xb,K,'same')

figk,axk = plt.subplots(1,1,figsize=(3,2))
axk.plot(t_kernel,K,'-r')
axk.set_xlabel('Time')
axk.set_ylabel('K(t)')

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(15,4))
ax[0].plot(t_data, rho1, color=(0.3,0.3,0.3,1), lw=1)
ax[0].set_ylabel('FR')
ax[1].plot(t_data, rho2, color='b', lw=1)
ax[1].set_ylabel('FR')
ax[1].set_xlabel('Time')
plt.show()

```
